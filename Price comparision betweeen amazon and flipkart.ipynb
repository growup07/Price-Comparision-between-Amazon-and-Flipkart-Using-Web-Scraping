# Importing Librariers

```python
from bs4 import BeautifulSoup as bs
import requests as r
import pandas as pd
import numpy as np
```

```python
flipkart_price = []
title=[]
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
```

# For Flipkart(Mobiles)

```python
name = input("enter the phone name ")
new_name = name.replace(',','')
u1="https://www.flipkart.com/search?q="  
u2=new_name
u3="&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_HistoryAutoSuggest_1_2_na_na_na&otracker1=AS_QueryStore_HistoryAutoSuggest_1_2_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=mobiles%7CMobiles&requestId=de8a6795-9b7e-4f76-a435-79726e79814a"
url = u1+u2+u3
print(url)
page = r.get(url,headers=headers)
soup = bs(page.content,"html.parser")
for x in soup.find_all('div', attrs={'class' : '_2kHMtA'}):
    flipkart_prices = x.find('div', attrs = {'class' : '_30jeq3 _1_WHN1'})
    if flipkart_prices is None:
        flipkart_price.append(np.NaN)
    else:
        flipkart_price.append(flipkart_prices.text)
        title_name = x.find('div', attrs = {'class' : '_4rR01T'})
    if title_name is None:
        title.append(np.NaN)
    else:
        title.append(title_name.text)
```

# Creating DataFrames For Flipkart

```python
df1 = pd.DataFrame({"title_name":title,"flipkart_prices":flipkart_price}).head(1)
```

```python
df1
```

# For Amazon(Mobiles)

```python
amazon_price = []
title = []
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
name = input("enter the phone name ")
new_name = name.replace(',','')
u4="https://www.amazon.in/s?k="
u5=new_name
u6="&rh=n%3A1389401031&ref=nb_sb_noss"
url1=u4+u5+u6
print(url1)
page1 = r.get(url1,headers=headers)
soup1 = bs(page1.content,"html.parser")
for x in soup1.find_all('div',attrs={"class":"s-card-container s-overflow-hidden aok-relative puis-include-content-margin puis s-latency-cf-section s-card-border"}):
        amazon_prices = x.find("span",attrs={"class":"a-price-whole"})
        if amazon_prices is None:
            amazon_price.append(np.NaN)
        else:
             amazon_price.append(amazon_prices.text)
             title_name = x.find("span",attrs={"class":"a-size-medium a-color-base a-text-normal"})
        if title_name is None:
            title.append(np.NaN)
        else:
             title.append(title_name.text)
```
